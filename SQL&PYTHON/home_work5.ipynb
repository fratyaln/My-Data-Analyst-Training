```python
import pandas as pd
import seaborn as sns
```

```python
df_person=pd.read_excel("HR.xlsx",sheet_name="PERSON")
df_department=pd.read_excel("HR.xlsx",sheet_name="DEPARTMENTS")
```

```python
# 1. Retrieve all employees' first name, last name, and salary

df_person[["NAME_","SURNAME","SALARY"]]
```

```python
# 2. List only female employees (GENDER = 'K') with first name, last name, and date of birth
df_person[df_person["GENDER"]=="K"][["NAME_","SURNAME","BIRTHDATE"]]
```

```python
# 3. List names and hiring dates of employees who joined after 2017
df_person[df_person["INDATE"]>"2017-12-30"][["NAME_","INDATE"]]
```

```python
# 4. Add a new employee
new_employee = {
    'ID':1781,
    'CODE':1781,
    'TCNUMBER': 12345678901,
    'NAME_': 'Ali',
    'SURNAME': 'Veli',
    'GENDER': 'E',
    'BIRTHDATE': '1985-12-05',
    'INDATE': '2022-01-15',
    'OUTDATE': None,
    'DEPARTMENTID': 3,
    'POSITIONID': 40,
    'PARENTPOSITIONID':42,
    'MANAGERID':None,
    'TELNR':'(0322) 4235916',
    'SALARY': 6000
}
# Convert the dictionary into a single-row DataFrame
new_employee_df = pd.DataFrame([new_employee])

# Then concatenate it with your existing DataFrame
df_person = pd.concat([df_person, new_employee_df], ignore_index=True)
df_person.tail()
```

```python
# 5. Update Ferhat CINAR'S salary to 6000 
df_person.loc[(df_person["NAME_"]=="Ferhat")&(df_person["SURNAME"]=="CINAR"),"SALARY"]=6000
```

```python
#6. Delete (drop) the employee named Deniz Eravcı from the DataFrame.
df_person = df_person[(df_person['NAME_'] != 'Deniz')&(df_person["SURNAME"]!="Eravcı")]
```

```python
# 7. List employees born before 1960
df_person[df_person["BIRTHDATE"]<"19600101"]
```

```python
# 8. List Employees born before 1960 and salary > 5000
df_person[(df_person["BIRTHDATE"]<"19600101")&(df_person["SALARY"]>5000)]
```

```python
# 9. Department ID is 4 OR salary > 5500
df_person[(df_person['DEPARTMENTID'] == 4) | (df_person['SALARY'] > 5500)]
```

```python
# 10. Increase salaries by 10% for employees with no exit date (OUTDATE is NULL/NaN)
df_person.loc[df_person['OUTDATE'].isna(), 'SALARY'] *= 1.
```

```python
# 11. Remove employees who were hired before 2015 and have a salary less than 5000.
df_person['INDATE'] = pd.to_datetime(df_person['INDATE'])  # Ensure 'INDATE' is in datetime format
df_person = df_person[~((df_person['INDATE'] < "2015-01-01") & (df_person['SALARY'] < 5000))]
```
