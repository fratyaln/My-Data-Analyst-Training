**Task 17: Load the Tips dataset from the Seaborn library.**

```python
tp=sns.load_dataset("tips")
tp.head()
```

**Task 18: Find the sum, minimum, maximum, and average (mean) of total_bill for each time category (Dinner, Lunch).**

```python
tp.groupby("time").agg({"total_bill":["sum","min","max","mean"]})
```

**Task 19: Find the sum, minimum, maximum, and average (mean) of total_bill for each day and time combination.**

```python
tp.groupby(["time","day"]).agg({"total_bill":["sum","min","max","mean"]})
```

**Task 20: Find the sum, minimum, maximum, and average of total_bill and tip values for female customers during Lunch time, grouped by day.**

```python

tp[(tp["sex"]=="Female")&(tp["time"]=="Lunch")].groupby("day").agg({"total_bill":["sum","min","max","mean"]})
```

**Task 21: What is the average of orders where size is less than 3 and total_bill is greater than 10? (Use loc)**

```python
tp.loc[(tp.loc[:,"size"]<3)&(tp.loc[:,"total_bill"]>10),"total_bill"].mean()
```

**Task 22: Create a new variable called total_bill_tip_sum. This variable should show the sum of total_bill and tip for each customer.**

```python
tp["total_bill_tip_sum"]=tp["total_bill"]+tp["tip"]
tp.head()
```

**Task 23: Sort the dataset by total_bill_tip_sum in descending order and assign the top 30 customers to a new DataFrame.**

```python
tp.sort_values(by="total_bill_tip_sum", ascending=False).head()
```
